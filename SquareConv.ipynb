```python
import os
import sys

import cv2
import numpy as np
```

```python
GT_PATH = './ASSR/gt/subset/'
TRAIN_PATH = './ASSR/images/subset/'

GT_SQUARE = './ASSR/gt/square/'
TRAIN_SQUARE = './ASSR/images/square/'

# Create square directories if they don't exist
if not os.path.exists(GT_SQUARE):
    os.makedirs(GT_SQUARE)

if not os.path.exists(TRAIN_SQUARE):
    os.makedirs(TRAIN_SQUARE)
```

```python
# List all images in the GT_PATH
gt_images = os.listdir(GT_PATH)
train_images = os.listdir(TRAIN_PATH)

# Reshape all images to square
for im in gt_images:
    image = cv2.imread(GT_PATH + im)
    shape = image.shape

    if shape[0] != shape[1]:
        print('Reshaping image', im)
        image = cv2.resize(image, (shape[0], shape[0]))
        cv2.imwrite(GT_SQUARE + im, image)

for im in train_images:
    image = cv2.imread(TRAIN_PATH + im)
    shape = image.shape

    if shape[0] != shape[1]:
        print('Reshaping image', im)
        image = cv2.resize(image, (shape[0], shape[0]))
        cv2.imwrite(TRAIN_SQUARE + im, image)
```
